### Extracting the angular frequency

Recall that in the second midterm and subsequent weeks we developed a strategy for a variant of this problem which used the following steps:

1. Solve the differential equation for $Q(t)$ for initial value $Q_0$ (take the initial velocity to be zero);
2. Use `ipywidgets` to explore this solution in a figure and extract approximate values of the period $T$ as a function of $Q_0$. We found $T$ by getting the time from peak to peak or valley to valley;
3. Interpolate the approximate values of $T$;
4. Use in interpolated values of $T$ as a guess for a minima finding algorithm;
5. Convert the extracted values of $T$ to angular frequency by $\omega=\frac{2\pi}{T}$.
6. To deal with the stationary solutions (e.g. $Q_0 = 0$), we used analytic results for the angular frequency by taking using the formula $\omega = \sqrt{\frac{1}{M}\frac{d^2\Delta U}{d Q^2}}$.

```python
### 4

# Initialize Constants
K = 1.0  # Quadratic coefficient
D = 2.4  # Quartic coefficient
Phi = 1.0  # Sextic coefficient
M = 1.0  # Mass of the oscillator

# Define Potential Energy and its Derivative
def potential_energy(Q):
    return (1/2) * K * Q**2 - (1/4) * D * Q**4 + (1/6) * Phi * Q**6

def dU_dQ(Q):
    return K * Q - D * Q**3 + Phi * Q**5

# Define Equations of Motion
def equations_of_motion(t, y):
    Q, Q_dot = y
    Q_ddot = -dU_dQ(Q) / M
    return [Q_dot, Q_ddot]

# Integrate the Differential Equation
from scipy.integrate import solve_ivp

def solve_trajectory(Q0, t_span, t_eval):
    initial_conditions = [Q0, 0]  # Initial position Q0 and zero velocity
    solution = solve_ivp(
        equations_of_motion, t_span, initial_conditions, t_eval=t_eval, method="RK45"
    )
    return solution.t, solution.y[0]

# Find Peaks to Extract Period
def find_peaks(Q_t):
    from scipy.signal import find_peaks
    peaks, _ = find_peaks(Q_t)
    return peaks

# Calculate Angular Frequency from Period
def calculate_frequency(peaks, t):
    if len(peaks) > 1:
        periods = [t[peaks[i + 1]] - t[peaks[i]] for i in range(len(peaks) - 1)]
        T_avg = np.mean(periods)
        omega = 2 * np.pi / T_avg
        return omega
    return None

# Generate Plot of \omega vs Q0
def generate_frequency_plot(Q0_values, t_span, t_eval):
    frequencies = []
    for Q0 in Q0_values:
        t, Q_t = solve_trajectory(Q0, t_span, t_eval)
        peaks = find_peaks(Q_t)
        omega = calculate_frequency(peaks, t)
        frequencies.append(omega if omega else 0)

    # Plot Results
    plt.figure(figsize=(6, 5))
    plt.plot(Q0_values, frequencies, label="$\omega$ vs $Q_0$", color="blue")
    plt.xlabel("Initial Amplitude $Q_0$")
    plt.ylabel("Angular Frequency $\omega$")
    plt.title("Angular Frequency vs Initial Amplitude")
    plt.axhline(y=np.sqrt(K / M), color="gray", linestyle="--", label="Harmonic Frequency")
    plt.legend()
    plt.grid()
    plt.show()

# Execute Analysis
Q0_values = np.linspace(-1.7, 1.7, 50)  # Range of initial amplitudes
t_span = (0, 50)  # Time range for simulation
t_eval = np.linspace(0, 50, 1000)  # Time points for evaluation
generate_frequency_plot(Q0_values, t_span, t_eval)
```

### Plot of the potential energy

```python
import numpy as np
import matplotlib.pyplot as plt

# define potential function
def potential_energy(Q, K=1.0, D=2.4, Phi=1.0):
    return (1/2) * K * Q**2 - (1/4) * D * Q**4 + (1/6) * Phi * Q**6

# define mesh for plotting
Q_mesh = np.linspace(-2,2,100)
U_mesh = potential_energy(Q_mesh)

# construct figure
plt.figure(figsize=(5, 4))
plt.plot(Q_mesh, U_mesh, color="gray")
plt.xlabel("$Q$")
plt.ylabel("$\Delta U$")
plt.xlim(-2,2)
plt.ylim(-0.1,0.25)
plt.show()
```
